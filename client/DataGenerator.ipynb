```python
import csv
import random
import string
```

```python
def make_book(title, authors, publisher, publishedDate, categories):
    return [
        title,             # title
        None,              # description
        authors,           # authors
        None,              # image
        None,              # previewLink
        publisher,         # publisher
        publishedDate,     # publishedDate
        None,              # infoLink
        categories,        # categories
        None,              # ratingsCount
]

def make_review(id, title, score, text):
    return [
        id, # ID
        title, # TITLE
        None, # PRICE
        None, # USER_ID
        None, # PROFILE_NAME
        None, # HELPFULNESS
        score, # SCORE
        None, # TIME
        None, # SUMMARY
        text, # TEXT
    ]

def write_books(books, name):
    random.shuffle(books)
    with open(f"data/{name}.csv", "w") as f:
        writer = csv.writer(f)
        writer.writerow(["Title","description","authors","image","previewLink","publisher","publishedDate","infoLink","categories","ratingsCount"])
        for book in books:
            writer.writerow(book)

def write_reviews(reviews, name):
    random.shuffle(reviews)
    with open(f"data/{name}.csv", "w") as f:
        writer = csv.writer(f)
        writer.writerow(['Id','Title','Price','User_id','profileName','review/helpfulness','review/score','review/time','review/summary','review/text'])
        for review in reviews:
            writer.writerow(review)

# test
b = make_book("lalala", ["jose", "perez"], "hola", "2023", ["ficcion"])
r = make_review("asd123", 'hola', 123.2, "MALAZO")


def generate_random_string(length):
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))

def random_authors(n):
    return [generate_random_string(20) for _ in range(0,n)]
    
def random_categories(n):
    return [generate_random_string(10) for _ in range(0,n)]
```

```python
def make_q3_reviews(tit):
    return make_review(
        id=generate_random_string(25),
        title=tit,
        score=score,
        text=generate_random_string(500),
    )

books = []
reviews = []

best_books = {}
# 20 cumplen
for i in range(0, 20):
    tit = f'Libro que cumple Num({i})'
    new_book = make_book(
        title=tit,
        authors= random_authors(random.randint(1,3)),
        publisher=generate_random_string(10),
        publishedDate=1990+random.randint(0,9),
        categories=random_categories(random.randint(1, 3)),
    )
    books.append(new_book)
    n_reviews = random.randint(500, 1000)
    scores = []
    for j in range(n_reviews):
        score = random.randint(1,5)
        scores.append(score)
        new_review = make_review(
            id=generate_random_string(25),
            title=tit,
            score=score,
            text=generate_random_string(500),
        )
        reviews.append(make_q3_reviews(tit))
    best_books[tit] = sum(scores)/len(scores)

# 10 que superan 500 pero no son del 90
for i in range(0, 10):
    tit = f'Libro que supera 500 pero no es del 90 Num({i})'
    new_book = make_book(
        title=tit,
        authors= random_authors(random.randint(1,3)),
        publisher=generate_random_string(10),
        publishedDate=2000+random.randint(0,9),
        categories=random_categories(random.randint(1, 3)),
    )
    books.append(new_book)
    n_reviews = random.randint(500, 1000)
    for j in range(n_reviews):
        reviews.append(make_q3_reviews(tit))

# 10 que no superan 500 y son del 90
for i in range(0, 10):
    tit = f'Libro que no supera 500 pero es del 90 Num({i})'
    new_book = make_book(
        title=tit,
        authors= random_authors(random.randint(1,3)),
        publisher=generate_random_string(10),
        publishedDate=1900+random.randint(0,9),
        categories=random_categories(random.randint(1, 3)),
    )
    books.append(new_book)
    n_reviews = random.randint(10, 499)
    for j in range(n_reviews):
        reviews.append(make_q3_reviews(tit))

# 10 que no superan 500 y no son del 90
for i in range(0, 10):
    tit = f'Libro que no supera 500 ni tampoco es del 90({i})'
    new_book = make_book(
        title=tit,
        authors= random_authors(random.randint(1,3)),
        publisher=generate_random_string(10),
        publishedDate=2000+random.randint(0,9),
        categories=random_categories(random.randint(1, 3)),
    )
    books.append(new_book)
    n_reviews = random.randint(10, 499)
    for j in range(n_reviews):
        reviews.append(make_q3_reviews(tit))

# los 10 mejores sobre los que cumplen son:
print(best_books)
```
